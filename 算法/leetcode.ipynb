```javascript
{
    function ListNode(val){
        this.val = val;
        this.next = null;
    }

    // head = 1-> 4 -> 3 -> 2 -> 5 -> 2 , x = 3
    let arr = [1,4,3,2,5,2];
    let curr = null;
    let head = null;
    arr.forEach(item=>{
        if(head==null){
            head = new ListNode(item);
            curr = head;
        }else{
            let temp = new ListNode(item);
            curr.next = temp;
            curr = temp;
        }
    })
    console.log('arr is=',arr);
    console.log('head is=',linkNodePrint(head));
    
    function linkNodePrint(obj){
        let res = [];
        while(obj){
            res.push(obj.val);
            obj = obj.next;
        }
        return res.join('->');
    }

    var partition = function(head, x){
        let curr = null;
        let temp = head;
        while(temp&&curr == null){
            if(temp.val==x){
                curr = temp;
            }
            temp = temp.next;
        }
        temp = head;
        let last = curr;
        while(last){
            let _obj = last.next;
            if(_obj.val<x){
                last.next = _obj.next;
                let _prev = temp;
                let flag = true;
                while(temp&&temp!=curr&&flag){
                    if(_obj.val>temp.val){
                        _obj.next = temp;
                        _prev.next = _obj;
                        flag = false;
                    }else{
                        _prev = temp;
                        temp = temp.next;
                    }
                }
            }else{
                last = last.next;
            }
        }
    }
    partition(head,3);
    console.log('now is=',linkNodePrint(head));
}
```
